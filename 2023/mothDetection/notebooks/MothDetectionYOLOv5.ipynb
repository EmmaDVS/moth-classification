# Moth detection with YOLOv5

To make the task of identifying moth species easier, images neatly cropped around the object of interest should be fed to a classification model. That means, do not just crop a square out of the middle of an image, because you might cut your object in half doing that. Instead, another model can be trained to determine if and where the object of interest is located in the image. It could even detect multiple objects in the same image. This task is called 'object detection'.

In this notebook, a pre-trained YOLOv5 model will be fine-tuned to detect moths of all shapes and sizes.
